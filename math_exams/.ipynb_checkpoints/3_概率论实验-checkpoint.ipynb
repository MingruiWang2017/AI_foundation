# 概率论实验

概率论是研究随机现象数量规律的数学分支。随机现象是相对于决定性现象而言的，在一定条件下必然发生某一结果的现象称为决定性现象。

概率论是用来描述不确定性的数学工具，很多数据挖掘算法都是通过样本的概率相关信息或推断来构建模型的。

```python
# 导入相关库
import numpy as np
import scipy as sp
```

### 6、二项式分布

服从二项式分布的随机变量X 表示在n次独立同分布的伯努利试验桩成功的次数，其中每一次一眼成功的概率为p。

```python
from scipy.stats import binom, norm, beta, expon
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# n,p 对应二项式分布中的事件成功次数及其概率，size 表示采样次数
binom_sim = binom.rvs(n=10, p=0.3, size=10000)
print('Data:', binom_sim, binom_sim.shape)
print('Mean: %g' % np.mean(binom_sim))
print('STD: %g' % np.std(binom_sim, ddof=1))
```

```python
# 生成直方图， x 指定每个bin（箱子）分布的数据，对应x轴， bins是总共有几根柱状图， 
# normed：值密度，也就是每个条状图的占比比例，默认是1
plt.hist(binom_sim, bins=10, normed=True)
plt.xlabel(('x'))
plt.show()
```

### 7、泊松分布

一个服从泊松分布的随机变量X，表示在具有比率参数$\lambda$ 的一段固定时间间隔内，时间发生的次数。参数 $\lambda$ 告诉你该事件发生的比率。随机变量 X的均值和方差都是 $\lambda$。

```python
# 产生10000 个符合lambda=2 的泊松分布的数
X = np.random.poisson(lam=2, size=10000)
```

```python
a = plt.hist(X, bins=15, normed=True, range=[0,15])
plt.grid(True)
plt.show()
```

### 8、正态分布
正态分布是一种连续分布，奇函数可以表示为在实线上的任何地方取值。正态分布由两个参数描述：分布的平均值$\mu$ 和标准差$\sigma$。

```python
mu = 0
sigma = 1
# 分布点采样
x = np.arange(-5, 5, 0.1)
# 生成符合mu，sigma 的正态分布（此处是标准正态分布）
y = norm.pdf(x, mu, sigma)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('density')
plt.show()
```

### 9、应用实例：图像加噪声

在深度学习等场景中，我们需要训练一个图像分类器，但是能拿到的图像不是很多，那有没有一种方法可以根据现有的图像生成一些新的图像出来呢？

增加的图像需要满足一些条件： 图像本身的内容不能发生根本性的改变，比如图像能看出来是一只狗的话，扩增出来的图像应该也是一只狗而不能只剩下一只毛茸茸的尾巴。

最直接的方式就是在原始图片上增加一些噪声，也就是使得图像变得粗糙同时又保持图像原始要表达的信息。

那么怎样给图像增加噪声呢？可以增加那些噪声呢？
我们接下来使用OpenCV来进行实验。

```python
import cv2
```

```python
# 载入原始图片
srcImage = cv2.imread('./lena.jpg')
print(srcImage.shape)
```

```python
# 显示图像
# cv2.namedWindow("Original image")  # 图像显示的窗口名称
cv2.imshow("original image", srcImage)
k = cv2.waitKey(0)
```

```python
# 灰度处理原始图片
grayImage = cv2.cvtColor(srcImage, cv2.COLOR_BGR2GRAY)
print(grayImage.shape)
cv2.imshow("gray image", grayImage)
k = cv2.waitKey(0)
```

#### 加入高斯噪声（均值：mean， 方差：var， 比例：percent)

```python
image = np.array(grayImage/255, dtype=float)
precent = 0.01  # 图像加入噪声的比例
num = int(precent * image.shape[0] * image.shape[1])

for i in range(num):
    temp1 = np.random.randint(image.shape[0])
    temp2 = np.random.randint(image.shape[1])
    
    mean = 0
    var = 0.04
    noise = np.random.normal(mean, var ** 0.5, 1)
    image[temp1][temp2] += noise

out = image
if out.min() < 0:
    low_clip = -1
else:
    low_clip = 0
out = np.clip(out, low_clip, 1)
gasuss_image = np.uint8(out * 255)
print(gasuss_image.shape)
cv2.imshow("gasuss image", gasuss_image)
k = cv2.waitKey(0)
```

#### 加入泊松噪声（泊松参数：scale， 比例：percent）

```python
image = np.array(grayImage, dtype=float)
percent = 0.001  # 图像加入噪声的比例
num = int(percent * image.shape[0] * image.shape[1])

for i in range(num):
    temp1 = np.random.randint(image.shape[0])
    temp2 = np.random.randint(image.shape[1])
    
    scale = 150
    noise = np.random.poisson(scale, 1)
    image[temp1][temp2] += noise
    
out = image
if out.min() < 0:
    low_clip = -1
else:
    low_clip = 0
out = np.clip(out, low_clip, 255)
expon_image = np.uint8(out)
print(expon_image.shape)
cv2.imshow("expon image", expon_image)
k = cv2.waitKey(0)
```

```python
# 设置属性，防止matplotlib图画中文乱码
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = [u'SimHei']
mpl.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(15,5))
plt.gray()  # 设置为灰度空间

plt.subplot(131)
plt.imshow(grayImage)
plt.title("原始灰度图")
plt.xticks([]),plt.yticks([])

plt.subplot(132)
plt.imshow(gasuss_image)
plt.title("加高斯噪声")
plt.xticks([]),plt.yticks([])

plt.subplot(133)
plt.imshow(expon_image)
plt.title("加泊松噪声")
plt.xticks([]),plt.yticks([])
```
